```python
import cv2 as cv
import numpy as np
from m2bk import *
from matplotlib import pyplot as plt
dataloader = DatasetHandler()
```

```python
f = dataloader.images_rgb[0]
h1 = cv.getGaussianKernel(21, 0)
h = h1.dot(h1.T)
g = cv.filter2D(f, ddepth=-1, kernel=h)

_, qlayout = plt.subplots(1, 2, figsize=(20, 20))
qlayout[0].imshow(f)
qlayout[1].imshow(g)
plt.show()
```
