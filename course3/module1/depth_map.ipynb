```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

imgL = cv.imread('aloeL.jpeg', cv.IMREAD_GRAYSCALE)
imgR = cv.imread('aloeR.jpeg', cv.IMREAD_GRAYSCALE)
# imgL = cv.pyrDown(cv.imread('aloeL.jpeg'))  # downscale images for faster processing
# imgR = cv.pyrDown(cv.imread('aloeR.jpeg'))

stereo = cv.StereoBM.create(numDisparities=128, blockSize=25)
disparity = stereo.compute(imgL,imgR)
plt.imshow(disparity,'gray')
plt.show()

```
