```python
import numpy as np 
from numpy.linalg import inv

X0_hat = np.array([[0, 5]]).T
P0_hat = np.diag([0.01, 1])
dt =0.5
u0 = -2
y1 = 2.2

F = np.identity(2)
F[0, 1] = dt
G = np.zeros((2, 1))
G[1, 0] = dt
H = np.array([[1, 0]])

# Measurement noise covariance
R0 = np.array([[0.05]])
# Process noise covariance 
Q0 = 0.1 * np.identity(2)

# Prediction step 
X1_check = np.dot(F, X0_hat) + G.dot(u0)
P1_check = np.dot(F, P0_hat).dot(F.T) + Q0


# Mesurement update step 
K = np.dot(P1_check, H.T).dot( inv(H.dot(P1_check).dot(H.T) + R0 ) )
X1_hat = X1_check + K.dot(y1 - H.dot(X1_check) )
P1_hat =  (np.identity(2) - K.dot(H)).dot(P1_check)

print("X1_hat", X1_hat)
print("P1_hat", P1_hat)
```
